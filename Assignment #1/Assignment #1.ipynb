Import all required libraries to execute this code

```python
import csv
import collections
from collections import defaultdict
import heapq
import matplotlib.pyplot as plt 
import numpy as np
import textwrap
from tabulate import tabulate
%matplotlib inline
```

Read csv data file and place only "Organization Name" and "Views Per Month" in list

```python
with open('datagovdatasetsviewmetrics.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    source = []
    for row in reader:
        org = row['Organization Name']
        view = row['Views per Month']
        source.append([org, view])
```

Create dictionary to easily aggregate organizations

```python
d = defaultdict(list)
for k, v in source:
    d[k].append(v)
```

Aggregate organizations and total the views for each organization

```python
agg_d = {k:sum(map(int, v)) for k, v in d.items()}
```

Create a function to find the top ten organizations with the highest number of views (returns a list)

```python
def dict_nlargest(d,n):
    return heapq.nlargest(n ,d, key = lambda k: d[k])
```

Identify the top 10 organizations with the highest views and return a list

```python
orgs = dict_nlargest(agg_d, 10)
```

Take top ten organizations, find their values (views), and add it to the 'views' list

```python
views = []
for o in range(0, len(orgs)):
    value = agg_d[orgs[o]]
    views.append(value)
```

Create a 2-D list ([org, views]) and print a table

```python
table = []
for i in range(0, len(orgs)):
    table.append([orgs[i], views[i]])
print tabulate(table, headers = ["Organization", "Total Views"], tablefmt= "simple")
```

Create and print a bar chart

```python
y = views
N = len(y)
x = np.arange(N)
width = 1
labels = [textwrap.fill(text,15) for text in orgs]
fig = plt.figure(num=1, figsize=(30,10))
plt.bar(x, y, width)
plt.xticks(x + width/2.0, labels)
plt.xlabel('Total Views')
plt.ylabel('Organization')
plt.title('Top Ten Organizations')
plt.show()
```
